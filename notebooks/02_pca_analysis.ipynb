```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df= pd.read_csv('C:\\Users\\MR\\Documents\\Heart_Disease_Project\\data\\preprocessed_data.csv')
df
```

```python
x= df.drop('condition', axis=1)
y= df['condition']
```

```python
pca = PCA(n_components=16)
x_pca= pca.fit_transform(x)
pca_df= pd.DataFrame(data=x_pca, columns=['PC1', 'PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13','PC14','PC15','PC16'])
print("PCA Variance =",pca.explained_variance_ratio_ )
pca_df 
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(explained_variance) + 1),
         cumulative_variance,
         marker="o", linestyle="-", color="b")
plt.axhline(y=0.98, color="r", linestyle="--")  
plt.title("Cumulative Explained Variance by PCA Components")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.scatter(pca_df['PC2'], pca_df['PC3'])
plt.xlabel('Principal Component 2')
plt.ylabel('Principal Component 3')
plt.title('PCA Scatter Plot (Principal Component 2 vs Principal Component 3)')
plt.grid(True)
plt.show()
```

```python
pca_prepossed_data= pd.concat([pca_df, y], axis=1)
pca_prepossed_data 
```

```python
pca_prepossed_data.to_csv('C:\\Users\\MR\\Documents\\Heart_Disease_Project\\data\\pca_preprocessed_data.csv', index=False)
```
